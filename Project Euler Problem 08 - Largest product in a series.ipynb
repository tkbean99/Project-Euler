```python
# We will define a function find the the product of a list of 13 numbers
def Product(List):
    resultant_product = 1
    for ele in List:
        resultant_product = resultant_product*ele
    return resultant_product
```

```python
# Now we will create a loop that does the following
# Step 1 : Start a loop that ends once the last set of consecutive numbers has been multiplied
# Step 2: Create a list, List_of_4, containing the first 4 digits. Find the resultant product of the first 4 digits using the earlier defined function
# Step 3: Temporarily store that product in a variable
# Step 4: Remove the first number in the list, add the  
x=0
y=4
List_of_4=[]
# This list 
for digit in Number_List[x:y]:
    List_of_4.append(digit)
Product_of_4 = Product(List_of_4)
Greatest_Product = Product_of_4
while y<1000:
    List_of_4.clear()
    x += 1
    y += 1
    for digit in Number_List[x:y]:
        List_of_4.append(digit) 
    Product_of_4 = Product(List_of_4)
    if Product_of_4 > Greatest_Product:
        Greatest_Product = Product_of_4
print(Greatest_Product)
```
