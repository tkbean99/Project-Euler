# Project Euler Problem 12 - Highly divisible triangular number

```python
'''
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:
1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
Let us list the factors of the first seven triangle numbers:
 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.
What is the value of the first triangle number to have over five hundred divisors?
'''
```

```python
# Let's create a function to find factors
def Factoriser(test_number):
    Factors=[1,test_number]
    x=2
    y = test_number/x
    while x<y:
        if test_number%x==0:
            Factors.append(x)
            y = int(test_number/x)
            Factors.append(y)
            x+=1
        else:
            x+=1
    return len(Factors)
```

```python
# Let's test our function
Factoriser(28)
```

```python
# I am going to create a loop that does the following
# Step 1: Generate a triangle number, put it in a list
# Step 2: Test the triangle number, find the number of factors
# Step 3: Check if the number of factors is equal to or exceeds 500
# Step 4: Repeat
# I am going to test for up to or more than 5 factors first before doing 500
# Create variable accounting for how many triangle numbers we have looked through
Series_Number=1
# Create an indicator to tell us when the answer we seek has been found
Answer_Indicator=False
while Answer_Indicator == False:
    # Create list to help create new triangle numbers
    Triangle_Number_Layers=[n for n in range (1,Series_Number+1)]
    # Create variable to store triangle number requiring testing
    Triangle_Number=sum(Triangle_Number_Layers)
    # Do an 'if' statement to check if the triangle number has more than 5 factors
    if Factoriser(Triangle_Number)>=5:
        print(Triangle_Number)
        Answer_Indicator=True
    # Otherwise, move on to the next triangle number
    elif Factoriser(Triangle_Number)<5:
        Series_Number += 1
        Answer_Indicator=False
```

```python
# Now we try to look for a triangle number with more than or equal to 500 factors
Series_Number=1
Answer_Indicator=False
while Answer_Indicator == False:
    Triangle_Number_Layers=[n for n in range (1,Series_Number+1)]
    Triangle_Number=sum(Triangle_Number_Layers)
    # Do an 'if' statement to check if the triangle number has more than 500 factors
    if Factoriser(Triangle_Number)>=500:
        print(Triangle_Number)
        Answer_Indicator=True
    # Otherwise, move on to the next triangle number
    elif Factoriser(Triangle_Number)<500:
        Series_Number += 1
        Answer_Indicator=False
```
